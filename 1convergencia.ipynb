```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a) Estime I usando Monte Carlo com amostragem uniforme.

# Número total de amostras
N = 100000

valor_real = 2 * np.log(2) - 1

# Execução da Simulação de Monte Carlo (Item a)
u = np.random.uniform(0, 1, N)  # Passo de Amostragem Uniforme
g_u = np.log(1 + u)              # Passo de Calcular g(u)
estimativa_final = np.mean(g_u)  # Passo de Média (Estimativa de I)

print(f"Valor Analítico: {valor_real:.6f}")
print(f"Estimativa Monte Carlo (N={N}): {estimativa_final:.6f}")


# Preparação do Gráfico de Convergência
# Calculamos a média acumulada para ver a evolução da estimativa
medias_acumuladas = np.cumsum(g_u) / np.arange(1, N + 1)

plt.figure(figsize=(9, 5))
plt.plot(medias_acumuladas, color='blue', label='Estimativa de Monte Carlo')
plt.axhline(y=valor_real, color='red', linestyle='--', label=f'Valor Real ({valor_real:.4f})')

plt.title('Questão 1(a): Estimativa de I via Monte Carlo')
plt.xlabel('Número de Amostras (n)')
plt.ylabel('Valor da Integral')
plt.legend()
plt.grid(True)
plt.xscale('log')
plt.show()
```

```python
# b) Execute simulações com N = 10², 10³, 10⁴, 10⁵.

# Valor Exato
valor_real = 2 * np.log(2) - 1

# Simulação (Item b)
Ns = [10**2, 10**3, 10**4, 10**5]

estimativas = []
erros_absolutos = []

for N in Ns:
    # Geração de amostras uniformes
    u = np.random.uniform(0, 1, N)
    # Cálculo da média de Estimativa de Monte Carlo 
    estimativa = np.mean(np.log(1 + u))
    estimativas.append(estimativa)
    
    # Cálculo do erro absoluto
    erros_absolutos.append(np.abs(estimativa - valor_real))

# Por resultados em tabela
print(f"{'N':>10} | {'Estimativa':>15} | {'Erro Absoluto':>15}")
for i in range(len(Ns)):
    print(f"{Ns[i]:10d} | {estimativas[i]:15.6f} | {erros_absolutos[i]:15.6e}")

#  Gráfico de Erro Absoluto vs N
plt.figure(figsize=(8, 4))
plt.plot(Ns, erros_absolutos, marker='o', color='red', linestyle='--')
plt.xscale('log')  #  visualizar a ordem de grandeza de N
plt.yscale('log')  # para o erro
plt.xlabel('Número de Amostras (N)')
plt.ylabel('Erro Absoluto')
plt.title('Convergência de Monte Carlo: Erro Absoluto vs N')
plt.grid(True, which="both", ls="-")
plt.show()

```

```python
# c) Plote o erro absoluto em função de N.

# calcular a média da função g(x)=ln(1+x) e plotar a evolução do erro absoluto em função de N.

# Valor de exato de referência
valor_real = 2 * np.log(2) - 1

# Simulação de (N)
Ns = [10**2, 10**3, 10**4, 10**5]

erros_absolutos = []

# Loop para executar a simulação para cada N
for N in Ns:
    #N amostras uniformes entre 0 e 1
    u = np.random.uniform(0, 1, N)
    
    # Calcular a estimativa via média de g(u)
    estimativa = np.mean(np.log(1 + u))
    
    # Calcular o erro absoluto
    erro = np.abs(estimativa - valor_real)
    erros_absolutos.append(erro)
    print(f"N = {N:6d} | Estimativa = {estimativa:.6f} | Erro = {erro:.6e}")

#Gráfico
plt.figure(figsize=(8, 4))
plt.plot(Ns, erros_absolutos, marker='s', linestyle='-', color='red', linewidth=2)
plt.xscale('log')
plt.yscale('log')
plt.title('Questão 1(c): Erro Absoluto vs Número de Amostras (N)')
plt.xlabel('Número de Amostras (N)')
plt.ylabel('Erro Absoluto (Escala Log)')
plt.grid(True, which="both", ls="--", alpha=0.7)
plt.show()
```
