```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#a) Utilize simulação de Monte Carlo para estimar a probabilidade de que a soma dos
# resultados seja estritamente maior que 9

# Parâmetros
N_simulacoes = 100000
sucessos = 0
evolucao = []

# Processo de Monte Carlo [Ref: 219]
for i in range(1, N_simulacoes + 1):
    # Lançamento de dois dados equilibrados
    dado1 = np.random.randint(1, 7)
    dado2 = np.random.randint(1, 7)
    
    if (dado1 + dado2) > 9:
        sucessos += 1
    
    if i % 500 == 0:
        evolucao.append(sucessos / i)

print(f"Probabilidade Estimada: {sucessos/N_simulacoes:.4f}")


# Geração do Gráfico
plt.figure(figsize=(8, 4))
plt.plot(range(500, N_simulacoes + 1, 500), evolucao, label='Estimativa MC')
plt.axhline(y=0.1667, color='red', linestyle='--', label='Valor Teórico (1/6)')
plt.title('Convergência de Monte Carlo - Soma de Dados > 9')
plt.xlabel('Número de Simulações (N)')
plt.ylabel('Probabilidade')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#b) Compare o valor obtido com o valor teórico


# Parâmetros
N_simulacoes = 100000
valor_teorico = 1/6
sucessos = 0
evolucao_estimativa = []

# Processo de Monte Carlo [Ref: 222]
for i in range(1, N_simulacoes + 1):
    # Lançamento simulado de dois dados (1 a 6)
    dado1 = np.random.randint(1, 7)
    dado2 = np.random.randint(1, 7)
    
    if (dado1 + dado2) > 9:
        sucessos += 1
    
    # Armazenar evolução para o gráfico a cada 500 passos
    if i % 500 == 0:
        evolucao_estimativa.append(sucessos / i)

print(f"Probabilidade Estimada final: {sucessos/N_simulacoes:.4f}")


# Geração do Gráfico
plt.figure(figsize=(8, 4))
plt.plot(range(500, N_simulacoes + 1, 500), evolucao_estimativa, color='blue', label='Estimativa MC')
plt.axhline(y=valor_teorico, color='red', linestyle='--', label=f'Valor Teórico ({valor_teorico:.4f})')
plt.title('Convergência de Monte Carlo - Soma de Dois Dados > 9')
plt.xlabel('Número de Simulações (N)')
plt.ylabel('Probabilidade Estimada')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

```python
#c) Discuta como o numero de simulações afeta a precisão do resultado Passos para a inversão

# Parâmetros
N_max = 100000
valor_teorico = 1/6
sucessos = 0
historico_estimativas = []
eixo_x = []

# Simulação de Monte Carlo
for n in range(1, N_max + 1):
    dado1 = np.random.randint(1, 7)
    dado2 = np.random.randint(1, 7)
    
    if (dado1 + dado2) > 9:
        sucessos += 1
    
    # Registra a estimativa em intervalos para o gráfico
    if n % 200 == 0:
        historico_estimativas.append(sucessos / n)
        eixo_x.append(n)
print(f"Probabilidade Estimada final: {sucessos/N_max:.4f}")
print(f"Valor Teórico: {valor_teorico:.4f}")

# Gráfico de Simulação (N)
plt.figure(figsize=(8, 4))
plt.plot(eixo_x, historico_estimativas, color='blue', label='Estimativa Simulação')
plt.axhline(y=valor_teorico, color='red', linestyle='--', label=f'Valor Teórico ({valor_teorico:.4f})')
plt.title('Efeito de N na Precisão (Convergência de Monte Carlo)')
plt.xlabel('Número de Simulações (N)')
plt.ylabel('Probabilidade Estimada')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```
